**7.2** Given `A = [[2,1],[1,2]]`, compute determinant, inverse, and verify `A @ A_inv ≈ I`.

```python
A=np.array([[2,1], [1,2]])
det_A=np.linalg.det(A)
A_inv=np.linalg.inv(A)
is_identity=np.allclose(A@A_inv, np.eye(2))
print(f"Determinant: {det_A: .4f}")
print("\nInverse Matrix:")
print(A_inv)
print(f"\nIs it Identity Matrix? {is_identity}")
```

**7.3** For `A` above, compute eigenvalues and eigenvectors, and verify `A @ v ≈ λ v`.

```python
eigenvalues, eigenvectors=np.linalg.eig(A)

lam0=eigenvalues[0]
v0=eigenvectors[:,0]

left_side=A@v0
right_side=lam0*v0
is_verified=np.allclose(left_side, right_side)

print(f"Eigenvalues: {eigenvalues}")
print("\nEigenvectors Matrix:")
print(eigenvectors)
print(f"\nDoes A@v0 ≈ λ0*v0? {is_verified}")
```

**7.4** Solve the linear system `3x + 2y = 7` and `2x + 5y = 8`.

**9.4 A/B Testing**: Variant A converts 48/500 users; Variant B converts 62/500. Compute conversion rates and the **absolute lift** (B−A). Simulate 10,000 bootstrap samples of the lift.

```python
np.random.seed(123)

# Məlumatları daxil edirik:
n_A=500
n_B=500
success_A=48
success_B=62
simulate=10000

# İlk konversiya dərəcələri və mütləq artımı tapırıq:
rate_A=success_A/n_A
rate_B=success_B/n_B
abs_lift=rate_B-rate_A

# Binomial paylanmadan istifadə edib uğurlu istifadəçi saylarını generasiya edirik:
boot_scs_A=np.random.binomial(n=n_A, p=rate_A, size=simulate)
boot_scs_B=np.random.binomial(n=n_B, p=rate_B, size=simulate)

# Yeni konversiya dərəcələrini tapırıq:
boot_rate_A=boot_scs_A/n_A
boot_rate_B=boot_scs_B/n_B

# 10000 ədəd lift-i hesablayırıq
boot_lifts=boot_rate_B-boot_rate_A

# Print hissələrin qeyd edirik:
print("Initial results of the A/B test")
print("-"*50)
print(f"Rate A: {rate_A*100: .1f}%")
print(f"Rate_B: {rate_B*100: .1f}%")
print("-"*50)

print(f"{simulate} Result of the bootstrape simulation:")
print("-"*50)
print(f"Simulated moderate growth: {np.mean(boot_lifts)*100: .2f}%")
print(f"STD : {np.std(boot_lifts): .4f}")
print("*"*50)

```

**9.5 Sports Analytics**: Given player scores over 8 games, compute moving average with window size 3 using vectorized operations (no loops). *(Generate random integers 0–30.)*

```python
np.random.seed(123)

scores=np.random.randint(0, 31, size=8)

window_size=3

filtr=np.ones(window_size)/window_size
moving_avg=np.convolve(scores, filtr, mode='valid')

print("Player scores:")
print("Games:", scores)
print("^"*50)

print(f"Moving Average (Window={window_size}):")
print("^"*50)
print("Results:", np.round(moving_avg,2))
print("_"*50)
```

**9.7 Logistics**: There are 5 warehouses with inventories `[40, 55, 20, 80, 60]` and 4 stores with demands `[30, 50, 40, 35]`. Using NumPy (not loops), compute how many units remain after fulfilling demands sequentially from warehouses.

```python
warehouses=np.array([40,55,20,80,60])
demands=np.array([30,50,40,35])

agg_demand=np.sum(demands)
cum_sum_warehouses=np.cumsum(warehouses)

remaining_load=np.minimum(warehouses, np.maximum(cum_sum_warehouses-agg_demand,0))
print("Final inventory in warehouses:", remaining_load)
```
